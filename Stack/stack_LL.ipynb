```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        
    def isempty(self):
        return self.top == None
            
    def peak(self):
        if self.isempty():
            return 'Stack is empty'
        else:
            return self.top.data
        
    def push(self,value):
        new_node = Node(value)
        new_node.next = self.top
        self.top = new_node
        
    def peek(self):
        if self.isempty():
            return f'Stack empty'
        else:
            return self.top
        
    def traverse(self):
        temp = self.top
        while temp != None:
            print(temp.data, end=' ')
            temp = temp.next
            
    def pop(self):
        if self.isempty():
            return f'Stack empty'
        else:
            data = self.top.data
            self.top = self.top.next
            return data
        
    def size(self):
        length=0
        temp = self.top
        if self.isempty():
            return f'Stack empty'
        while temp != None:
            temp = temp.next
            length += 1
        return length
    
```

```python
s = Stack()

s.push(3)
s.push(4)
s.push(5)
s.push(6)

s.traverse()
s.isempty()
s.size()
```

### Reverse the String.

```python
def reverse(text):
    s = Stack()
    
    for i in text:
        s.push(i)
        
    res =""   
    while not(s.isempty()):
        res += s.pop()
    
    return res

reverse('hello')
```

### Text Editor.

```python
def text_editor(text,pattern):
    undo_Stack = Stack()
    redu_Stack = Stack()
    res = ''
    for i in text:
        undo_Stack.push(i)
    pattern.lower()
    for j in pattern:
        if j == "u":
            redu_Stack.push(undo_Stack.pop())
        if j == 'r':
            undo_Stack.push(redu_Stack.pop())
    while not(undo_Stack.isempty()):
        res = undo_Stack.pop()+res
        
    return res

text_editor('hello','uur')
```

### Celebrity Problem

```python
L = [
    [0,0,1,1],
    [0,0,1,0],
    [0,0,0,0],
    [1,0,1,0]
]

def find_the_celebrity(L):
    s = Stack()
    for i in range(len(L)):
        s.push(i)
   
    while s.size() >= 2:
        i = s.pop()
        j = s.pop()
        
        if L[i][j] == 0:
            s.push(i)
        else:
            s.push(j)
    celeb = s.pop()
    for i in range(len(L)):
        if i != celeb:
            if L[celeb][i] == 1 or L[i][celeb] == 0:
                return -1
                
    return f'The Celebrity is {celeb}'
find_the_celebrity(L)
        
```
