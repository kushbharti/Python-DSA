```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        
    def isempty(self):
        return self.top == None
            
    def peak(self):
        if self.isempty():
            return 'Stack is empty'
        else:
            return self.top.data
        
    def push(self,value):
        new_node = Node(value)
        new_node.next = self.top
        self.top = new_node
        
    def peek(self):
        if self.isempty():
            return f'Stack empty'
        else:
            return self.top
        
    def traverse(self):
        temp = self.top
        while temp != None:
            print(temp.data, end=' ')
            temp = temp.next
            
    def pop(self):
        if self.isempty():
            return f'Stack empty'
        else:
            data = self.top.data
            self.top = self.top.next
            return data
    
    
```

```python
s = Stack()

s.push(3)
s.push(4)
s.push(5)
s.push(6)

s.traverse()
s.isempty()
```

### Reverse the String.

```python
def reverse(text):
    s = Stack()
    
    for i in text:
        s.push(i)
        
    res =""   
    while not(s.isempty()):
        res += s.pop()
    
    return res

reverse('hello')
```
