```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        
    def isempty(self):
        return self.top == None
            
    def peak(self):
        if self.isempty():
            return 'Stack is empty'
        else:
            return self.top.data
    def push(self,value):
        new_node = Node(value)
        new_node.next = self.top
        self.top = new_node
```
