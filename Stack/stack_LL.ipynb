```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        
    def isempty(self):
        return self.top == None
            
    def peak(self):
        if self.isempty():
            return 'Stack is empty'
        else:
            return self.top.data
        
    def push(self,value):
        new_node = Node(value)
        new_node.next = self.top
        self.top = new_node
        
    def traverse(self):
        temp = self.top
        while temp != None:
            print(temp.data)
            temp = temp.next
```

```python
s = Stack()
```

```python
s.push(12)
s.push(10)

s.traverse()
s.isempty()
```
