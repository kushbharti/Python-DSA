```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
class LinkedList:
    
    def __init__(self):
        
        # Empty Linked List
        self.head = None
        # No. of nodes in the LL
        self.n = 0
        
    def __len__(self):
        return self.n
        
    def insert_head(self,value):
        
        # New node
        new_node = Node(value)
        
        # Create connection
        new_node.next = self.head
        
        # Reassign head
        self.head = new_node
        
        # Increament n
        self.n += 1
        
        
    def __str__(self):
        curr = self.head
        result =''
        while curr != None:
            result = f'{result} {str(curr.data)} ->'
            curr = curr.next
        return result[:-2]
    
    
    def append(self,value):
         # New node
        new_node = Node(value)
        if self.head == None:
            # Empty
            self.head = new_node
            self.n += 1
            return 
    
        # Create connection
        curr = self.head
        
        while curr.next != None:
            curr = curr.next
        
        # you are at last node
        curr.next = new_node
       
        # Increament n
        self.n += 1
        
    def insert_after(self,after,value):
        new_node = Node(value)
        curr = self.head
        while curr != None:
            if curr.data == after:
                break
            curr = curr.next
        
        # case 1 (item found):-
        if curr != None:
            new_node.next = curr.next
            curr.next = new_node
            self.n += 1
        else:
            # case 2 (item not found):-
            return 'Item not found'
        
    def clear(self):
        self.head = None
        self.n = 0
       
    def delete_head(self):
        if self.head == None:
            return 'Empty Linked List'
        self.head = self.head.next
        self.n -= 1
        
    def pop(self):
        # Empty linked List
        if self.head == None:
            return 'Empty Linked List'
    
        curr = self.head
        
        # if there is only one item in the linked list
        if curr.next == None:
            # since there is only one node, i.e Head only
            return self.delete_head()
        
        while curr.next.next != None:
            curr = curr.next
        # curr -> 2nd last node
        curr.next = None
        self.n -= 1
    
    def remove(self,value):
        
        # If linked list is empty
        if self.head == value:
            return 'Empty linked List'
        
        # Removing head if the value matchs
        if self.head.data == value:
            return self.delete_head()
        
        curr = self.head
        while curr.next != None:
            if curr.next.data == value:
                break
            curr = curr.next
        
        # case 1 (item found):-
        if curr.next != None:
            curr.next = curr.next.next
            self.n -= 1
        else:
            # case 2 (item not found):-
            return 'Item not found'
    
    def seach(self,iteam):
        curr = self.head
        pos = 0
        
        while curr != None:
            if curr.data == iteam:
                return pos
            
            curr = curr.next
            pos += 1
        
        return 'Not Found'
    
    def __getitem__(self,index):
        curr = self.head
        pos = 0
        
        while curr != None:
            if pos == index:
                return curr.data
            curr = curr.next
            pos +=  1
            
        return 'IndexError'
        
        
        
            
                
```
